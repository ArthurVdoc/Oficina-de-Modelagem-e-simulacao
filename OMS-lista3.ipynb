**1)** *No texto, foi deduzido o modelo matem´atico para um sistema massa-mola-amortecedor no qual a mola e o amortecedor viscoso se encontravam em paralelo. Deduzir a EDO para um sistema massa-mola-amortecedor no qual a mola e o amortecedor se encontram em série. Quantos graus de liberdade tem esse sistema?*

---

**Resposta:**  
Agora temos:  

> massa -- mola -- amortecedor -- chão.  

Como os sistema está em série, a mesma força passa por todos os elementos  
Vamos tomar as variáveis $y(t)$ para a posição da massa, $x_{1}(t)$ para a deformação da mola e $x_{2}(t)$ para a deformação do amortecedor. Assim, temos que $y(t) = x_{1}(t) + x_{2}(t)$.  
  
Alem disso, temos que haverá a mesma força em ambos os componentes: $F = kx_{1}(t) = c\dot{x}_{2}(t)$  

Dessa forma, derivando e substituindo:  
$x_{2}​(t) = y(t) − x_{1}​(t)$  
$\dot{x}_{2}​(t) = \dot{y}(t) − \dot{x}_{1}​(t)$  
$c\dot{x}_{1} + kx_{1}= c\dot{y}$  

Agora para a massa:  
$f(t) = F + m\ddot{y}(t)$  
Mas sabemos que $F = kx_{1}$  
Então $m\ddot{y}(t) = f(t) - kx_{1}$

Assim obtemos o sistema acoplado:  
$
\begin{cases}
m\ddot{y}(t) = f(t) - kx_{1} \\
c\dot{x}_{1} + kx_{1}= c\dot{y}
\end{cases}
$  

Que, por fim, ao isolar x1 na equação da masssa, derivar mais uma vez a equação e substituir $\dot{x}_1$ na outra equação, obtemos o sistema de tercira ordem:  
$mc\dddot{y}(t) + mk\ddot{y}(t) + kc\dot{y}(t) = c\dot{f}(t) + kf(t)$  

Assim, como o sistema que obtivemos é de terceira ordem com duas variáveis independentes ($x_{1}$ e $y$), teremos **2 Graus de Liberdade**.





**2)** *Determine os modelos matemáticos para os sistemas apresentados nas Figuras 13, 14, 15
e 16.*

---

**Resposta:**  

##### Figura 13  
Para este sistema temos como variáveis o comprimento $L$, a massa $m$ e o angulo $\theta$.  
A partir delas, ao analisar a imagem traçamos as seguintes equações:
- O torque restaurador: $\tau = −m⋅g⋅L⋅sin(θ)$
- O momento de Inércia: $I = mL^2$ 
- 2ª lei de Newton para rotação: $\tau = I \ddot{\theta}$  

Substituindo, temos a EDO:  
$$mL^2 \ddot{\theta}(t) + mgL \sin(\theta) = 0$$


##### Figura 14  
Aqui temos um sistema de translação que pode ser deduzido a partir da 2ª lei de Newton, em que o somatório das forças é igual a massa do sistema multiplicado pela aceleração, ou seja:  
$\sum F = ma$  
em que:  
$a = \ddot{x}(t)$  
$\sum F = f(t) - c\dot{x}$ (força aplicada e amortecedor).  
  
Substituindo, temos a EDO:
$$m\ddot{x}(t) + c\dot{x}(t) = f(t)$$


##### Figura 15  
Assim como para a figura anterior vamos usar a 2ª lei de Newton, mas agora para sistemas de rotação, ou seja:  
$\sum \Tau = I\ddot{\theta}$.  
Como a barra nao tem massa, o momento de inercia sofre influência apenas da massa na extremidade:  
$I = mL^{2}$.  
Além disso, para o calculo dos torques, precisamos dos deslocamentos da mola, da massa e do amortecedor:
- Deslocamento da mola: $y_{k} = \frac{2}{3}L\theta$, pois ela está a uma distancia $a = \frac{2}{3}L$ do eixo.
- Deslocamento da massa e do amortecedor: $y_{m} = L\theta$, pois ambos estão na extremidade de $L$.  

Assim podemos calcular os torques:  
- Mola: $-(k \cdot \frac{2}{3}L \theta) \cdot \frac{2}{3}L = -\frac{4}{9} k L^2 \theta$
- Amortecedor: $-(c \cdot L \dot{\theta}) \cdot L = -cL^2 \dot{\theta}$  

Dessa forma, substituindo as equações dos torques e o momento de inercia na equção da 2ª lei de Newton, chegamos em:
$$mL^2 \ddot{\theta} + cL^2 \dot{\theta} + \frac{4}{9} k L^2 \theta = 0$$
Que, ao simplificar por $L^{2}$, finalmente temos a EDO:  
$$m \ddot{\theta}(t) + c \dot{\theta}(t) + \frac{4}{9} k \theta(t) = 0$$


##### Figura 16  
Neste sistema de polias, temos que o deslocamento da roldana ($y_{p}$) é a metado do deslocamento da massa ($x$), que esta sustentada por um cabo preso na outra extremidade. A partir disso, deduz-se as seguintes relações:
- Deslocamento da polia: $y_{p} = \frac{1}{2}x$
- Força na mola: $F_{k} = ky_{p} = \frac{1}{2}kx$
- Tensão no cabo ($T$): Como $F_{k} = 2T$ e $F_{k} = \frac{1}{2}kx$, então $T = \frac{kx}{4}$
- Estática para a massa $m$: $m\ddot{x}(t) + T = 0$
  
Substituindo $T$, temos a EDO:  
$$m\ddot{x}(t) + \frac{k}{4}x(t) = 0$$

**3)** *Considere o pêndulo apresentado na Figura 13 e o modelo matemético encontrado no
Exercício 2. Assuma que, para pequenas oscilações, sen (θ) ≈ θ em radianos. Linearize o
modelo matemético do pêndulo, transformando-o em uma EDO linear.*

---

**Resposta:** No item anterior, obtivemos a EDO $$mL^2 \ddot{\theta}(t) + mgL \sin(\theta) = 0$$ para o pêndulo da figura 13. Agora, para atender às exigências do exercício, vamos substituir $\sin(\theta)$ por \theta e dividir a equação toda por $mL^2$.  
Com isso, temos a EDO: $$\ddot{\theta}(t) + \frac{g}{L}\theta(t) = 0$$
